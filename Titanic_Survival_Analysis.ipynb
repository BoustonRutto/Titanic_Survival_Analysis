#### c). Feature Engineering
Creating new columns; Age Group and Family Size  
Age Group: Children to Senior Adults  
Family Size: Parents and Children, Siblings and Spouses

```python
# Creating a Family Size Variable
df['Family_Size']= df['SibSp']+ df['Parch']+ 1
```

```python
# Creating Age Group column with values to include all the age groups who were aboard
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 80], labels=['Child', 'Teen', 'Adult', 'Middle-aged', 'Senior'])
```

#### d) Percentages
Survival and Casualties Percentage by:  
* Gender
* Class
* Age Group

##### 1. Survival Percentage

```python
# Gender Survival
gender_survival=df.groupby('Sex')['Survived'].mean() * 100

# Class Survival
class_survival = df.groupby('Pclass')['Survived'].mean() * 100

#Age Group Survival
agegroup_survival= df.groupby('Age_Group', observed = True)['Survived'].mean() * 100

print("\n Survival percentage by Gender: \n", gender_survival)  
print("\n Survival percentage by Class: \n", class_survival) 
print("\n Survival Percentage by Age Group:\n", agegroup_survival)
```

##### 2. Casualties Percentage

```python
# Gender Fatalities
gender_casualties = df.groupby('Sex')['Survived'].apply(lambda x: (x==0).mean()) * 100

# Class Casualties
class_casualties= df.groupby('Pclass')['Survived']. apply(lambda x: (x==0).mean()) * 100

# Age Group Casualties
agegroup_casualties= df.groupby('Age_Group', observed= True)['Survived'].apply(lambda x:(x==0).mean()) * 100

print("\n Gender Casualties Percentage:\n", gender_casualties)
print("\n Class Casualties Percentage:\n", class_casualties)
print("\n Age Group Casualties Percentage:\n", agegroup_casualties)
```
